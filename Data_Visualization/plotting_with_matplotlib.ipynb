```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(6, 4))
gs = gridspec.GridSpec(2, 2)

ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.exp(x))
ax3.plot(x, np.log(x))
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(1, 20))
ax.set(
    xlim=(0, 40),
    ylim=(0, 40)
)
ax.grid(True, linestyle='--', alpha=0.3, color='g')
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], label='Data')

ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(['Zero', 'One', 'Two', 'Three', 'Four'], rotation=45)
ax.set_yticks([1, 2, 3, 4])
ax.set_yticklabels(['Low', 'Medium', 'High', 'Very High'])
ax.legend()
```

# Basic Plot Types:

```python
fig, ax = plt.subplots(figsize=(6, 4))

x = np.linspace(-4*np.pi, 4*np.pi, 500)

ax.plot(x, np.sin(x), label='sin(x)', color='r')
ax.plot(x, np.cos(x), label='cos(x)', color='g')

ax.plot(x, np.log(x), label='log(x)', color='b', linestyle='--', linewidth=2)

ax.grid(True, color='black', alpha=0.3)

ax.set(
    xlim=(-(np.pi * 4), (np.pi * 4)),
    ylim=(-1.5, 3),
    xlabel='x',
    ylabel='f(x)'
)
ax.legend()
```

```python
np.random.seed(42)
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)

fig, ax = plt.subplots(figsize=(6, 4))

ax.scatter(x, y)
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 35, 14, 27, 10]

fig, ax = plt.subplots(figsize=(7, 5))

ax.bar(categories, values)

for i, v in enumerate(values):
    ax.text(i, v, str(v), ha='center', va='bottom')
```
