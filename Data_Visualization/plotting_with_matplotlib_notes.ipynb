```python
# 2b.
# Scatter Plots and their customization:
np.random.seed(42)
x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)
colors = np.random.rand(50)
sizes = 300 * np.random.rand(50)

fig, ax = plt.subplots(figsize=(6, 3))

scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='coolwarm')

ax.grid()

# to set colorbar:
# Note: 'cmap=viridis' is the default for the colorbar.
fig.colorbar(scatter).set_label('Color value')

# to set a trend line:
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax.plot(x, p(x), c='red', label='Trend Line')

ax.legend()

fig
```

```python
# How to plot the trend line:
# To learn how to plot a trend line, we first need to understand how to construct a mathematical function using NumPy. 
# A trend line is essentially a mathematical function, and every function is defined by a set of coefficients that must be calculated from the data.
# This is done using `np.polyfit()`, which takes the x values, y values, and a third argument representing the degree of the function.
# For example, if we want a first-degree trend line (a straight line), we set the third argument to 1.
# The output of `np.polyfit()` is an array containing the coefficients of the fitted function.
# Next, `np.poly1d()`, which takes an argument z (the coefficients) is used to convert these coefficients into an actual callable function.
# Once the function is created, we simply plot it alongside the data to display the trend line in our visualization.
```

```python
# 3b.
# Bar charts:
# It is useful for categorical data.
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 35, 14, 27, 10]

fig, ax = plt.subplots(figsize=(6, 4))

ax.bar(categories, values)

ax.set(
    title='Basic Bar Chart',
    xlabel='Categories',
    ylabel='Values'
)

# To add value labels on top of each bar:
for i, v in enumerate(values):
    ax.text(i, v, str(v), ha='center', va='bottom')
    
fig

# If needed, learn more customization from the documetnation.
```

```python
# 4b.
# Histograms:
data_1 = np.random.normal(100, 20, 1000)
data_2 = np.random.normal(80, 30, 1000)

fig, ax = plt.subplots(figsize=(8, 4))

ax.hist(data_1, bins=30, alpha=0.7, label='Dataset 1', edgecolor='black')
ax.hist(data_2, bins=30, alpha=0.7, label='Dataset 2', edgecolor='black')

ax.set(
    title='Customized Histogram',
    xlabel='Values',
    ylabel='Frequency'
)

ax.legend()

# To add vertical lines (for example for mean, variance, etc):
ax.axvline(data_1.mean(), c='red', linestyle='dashed', linewidth=2, label='Mean 1')
ax.axvline(data_2.mean(), color='green', linestyle='dashed', linewidth=2, label='Mean 2')

# To add a KDE plot:
from scipy import stats

fig, ax = plt.subplots(figsize=(8, 4))

# Plot histogram and KDE
ax.hist(data_1, bins=30, density=True, alpha=0.7, edgecolor='black')
kde = stats.gaussian_kde(data_1)
x_range = np.linspace(data_1.min(), data_1.max(), 100)
ax.plot(x_range, kde(x_range), label='KDE')

ax.set_title('Histogram with KDE')
ax.set_xlabel('Values')
ax.set_ylabel('Density')
ax.legend()

fig
```
