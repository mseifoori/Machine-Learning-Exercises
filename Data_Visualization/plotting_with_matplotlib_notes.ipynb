```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1a.
# To see the result in jupyter:
%matplotlib inline
```

```python
# 2a.
# We have figure and axes as subplots:
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# 3a.
# The best way to make a plot:
# fig, ax = plt.subplots(rows, columns, figsize=(6, 4))
fig, axs = plt.subplots(2, 2, figsize=(6, 4))
```

```python
# Each plot can be called by ax and be shown by calling fig:
# After plotting on an ax, see the figure by calling fig:
x = np.linspace(0, 10, 100)
axs[0, 1].plot(x, np.sin(x))

fig
```

```python
# Example 1:
np.random.seed(42)
a = np.random.rand(50)
b = a + np.random.rand(50)

axs[0, 0].set(
    xlim=(0, 1),
    ylim=(0, 2),
    xlabel='x title',
    ylabel='y title'
)

axs[0, 0].grid(True, linestyle='--', alpha=0.3, color='green')
axs[0, 0].scatter(a, b)

fig
```

```python
# Example 2:
np.random.seed(42)
data_1 = pd.Series(1 + (10 - 1) * np.random.rand(100))

fig, ax = plt.subplots(figsize=(6, 4))
ax.set(
    xlim=(1, 5),
    ylim=(1, 10),
    xlabel='data 1',
    ylabel='TO BE FIELD',
)
ax.grid(True, alpha=0.7, color='purple')
ax.plot(data_1)

# Note: .plot() shows the whole plot, while .scatter() and etc needs "fig" to be called.
```

```python
# 4a.
# We can also customize the axis labels, known as ticks.
# For example, instead of using numbers, we can represent them semi-quantitatively, such as "one plus," "two plus," and so on, or categorize them as "low," "medium," and "high."
fig, ax = plt.subplots(figsize=(6, 4))
ax.set(
    title='Status',
    xlabel='Protein Conc.',
    ylabel='Status',
    yticks=[0, 1, 2, 3, 4, 5],
    yticklabels=['Negative', '1+', '2+', '3+', '4+', '5+'],
)
ax.grid(True, linestyle='--', alpha=0.3, color='green')
ax.plot([1, 2, 3, 4, 5], label='Protein Conc.')
ax.legend()
```
