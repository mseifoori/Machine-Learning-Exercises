```python
email_df['First Name'] = email_df['First Name'].astype('str')
email_df['Last Name'] = email_df['Last Name'].astype('str')
email_df['Full Name'] = email_df['First Name'].str.cat(email_df['Last Name'], sep= ' ')
email_df['Full Name']
```

```python
print(email_df.columns.tolist())
```

```python
email_df = email_df.loc[:, ['Index', 'User Id', 'First Name', 'Last Name', 'Full Name', 'Sex', 'Email', 'Phone', 'Date of birth', 'Job Title']]
```

```python
email_df['Sex'] = email_df['Sex'].str.replace(pat='Female', repl='F')
email_df['Sex'] = email_df['Sex'].str.replace(pat='Male', repl='M')
```

```python
email_df
```

```python
email_domain_series =  email_df['Email'].str.split('@').str[1]
email_domain_series.index = np.arange(1, len(email_domain_series) + 1)
email_domain_series
```
