```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Patient_ID': [1, 2, 3, 4, 5, 6, 7],
    'Age': [25, 30, np.nan, 45, 50, np.nan, 40],
    'Test Type': ['A', 'B', 'A', np.nan, 'C', 'B', 'A'],
    'Result': ['Positive', np.nan, 'Negative', 'Negative', np.nan, 'Positive', 'Positive'],
    'Score': [85, 90, np.nan, 75, 88, 92, np.nan]
})

df
```

```python
df.info()
```

```python
df['Result'].isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(0)
```

```python
df.fillna({
    'Age': df['Age'].mean(),
    'Test Type': 'Unknown',
    'Result': 'Unknown',
    'Score': df['Score'].mean()
})
```

```python
df.ffill()
```

```python
df.bfill()
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Test Type'] = df['Test Type'].fillna('Unknown')
df['Result'] = df['Result'].fillna('Unknown')
df['Score'] = df['Score'].bfill()
```

```python
df
```

```python
df.replace([-1, 'N/A', 'Unknown', 'unknown', 'missing'], np.nan)
```

```python
df.replace({
    'Age': {-1: np.nan},
    'Result': {'N/A': np.nan, 'Missing': 'Unknown'}
})
```

```python
for value in df['Age']:
    if int(value) < 40:
        df['Age'] = df['Age'].replace(value, np.mean(df['Age']))
    elif int(value) == -1:
        df['Age'] = df['Age'].replace(value, np.nan)
    else:
        df['Age'] = df['Age']
```
